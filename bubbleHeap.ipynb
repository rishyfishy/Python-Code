```python
def bubbleOnceDown(L,i):
    #if the right-most number exists and is bigger
    if (L[2*i+2] and L[2*i+2]>=L[2*i-1]):
        L[i-1],L[2*i] = L[2*i],L[i-1]   #Swap parent and right child
        i=2*i+1                         #make i the index of the right child+1


    #If the leftmost number is bigger
    else:
        L[i-1],L[2*i-1] = L[2*i-1],L[i-1]
        i=2*i
            
def bubbleDown(L,j):
    i=1
    #bubble down first value
    while (2*i<j):
        print(i)
        bubbleOnceDown(L,i) #i doesn't get modified


def heapSort(L):
    j=len(L)
    while (j):
        print(float(j))
        #swap first and last values
        L[0],L[-1]=L[-1],L[0]
        bubbleDown(L,j)
        j-=1


    

```
